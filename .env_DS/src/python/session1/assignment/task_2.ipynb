```python

#question_1

email = input ("Enter Email address")

def validation (email) :
    if email.count("@") == 1 :
        at_index = email.index("@")
        return email[at_index:].count(".") > 0 
    else:
        return False

def extract_username (email) :
    return email [ : email.index("@")]

def extract_domain (email) :
    return email [ email.index("@")+1 : email.rindex(".")]

def extract_domain_end (email) :
    return email [ email.rindex(".")+1 : ]

def extract_domain_end_category (domain_end) :
    if domain_end == "com" :
        return "Commercial Domain"
    if domain_end == "edu" :
        return "Educational Domain"
    else :
        return "Other Domain"


if validation(email) :

    print("username is : ",extract_username (email))
    print("domain is : ",extract_domain (email))
    print("domain_end is : ",extract_domain_end (email))
    print("domain_end_category is : ",extract_domain_end_category (extract_domain_end (email)))
    
else :
    print("not valid Email")
```
