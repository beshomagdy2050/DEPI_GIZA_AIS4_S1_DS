# Write a Python program to remove an item from a given set

## Example
my_set: {1, 2, 3, 4, 5}

Output: {1, 2, 4, 5}

```python
# Write your answer here
list_1 = [1,2,3,4,5]
x = int(input("enter number to remove"))
if x in list_1 :
    list_1.remove(x)
print(list_1)
```

# Write a Python program to removes multiple items from a given set

## Example
my_set: {1, 2, 3, 4, 5}

items to be removed: {1, 4}

Output: {2, 3, 5}

```python
# Write your answer here
list_2 = [1,2,3,4,5]
y = True
while y :
    number = int(input("enter number to remove and press enter \n to exit enter 00 "))
    if number in list_2 :
        list_2.remove(number)
    if number == 00 :
        y = False
print(list_2)
```

# Write a Python program to find the index of an item in a tuple.

my_tuple: ('apple', 'banana', 'orange', 'grape', 'banana')

```python
# Write your answer here
tuple_1 = ('apple','banana','orange','grape','banana')
input_1 = input("enter any item to get index")
for i in range (len(tuple_1)) :
    if tuple_1[i] == input_1 :
        print(i)
```

## Example
(factorial of 5) = 5 * 4 * 3 * 2 * 1 = 120

```python
# Write your answer here
import math
input_3 = int(input("enter number to get factorial"))
print(math.factorial(input_3))

```

```python
## another solution 
def fact (number) :
    factorial = 1
    counter = 1
    while counter <= number :
        factorial *= counter
        counter +=1
    return factorial
input_4 = int(input("enter number to get factorial"))
print(fact(input_4))
```

# Create a function that Calculates the Sum of Even Numbers in a List
## Example
numbers: [1, 2, 3, 4, 5, 6]

Output: 12

```python
# Write your answer here
def sum_of_even (li) :
    sum = 0
    for i in li :
        if i % 2 == 0 :
            sum = sum + i
    return sum
list_4 = [1,2,3,4,5,6]
print(sum_of_even(list_4))
```

# Create a function that takes a list as input and return reversed list

## Example
lst: [1, 2, 3]

Output: [3, 2, 1]

```python
# Write your answer here
def reverse_list (lis) :
    x = []
    i = len(lis)-1
    while i >= 0 :
        x.append(lis[i])
        i -= 1
    return x

list_4 = [1,2,3,4,5,6]
print(reverse_list(list_4))
```
