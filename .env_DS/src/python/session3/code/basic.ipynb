```python
import os 

def make_dir (path):
    if not os.path.exists(path):
        os.makedirs(path)
    for i in range (20) :
        file_path = os.path.join(path,"Dir_"+str(i))
        if not os.path.exists(file_path):
            os.makedirs(file_path)
    print(path)


directory = "./test"
make_dir(directory)
```

```python
"""
THIS CLASS

"""

class StudentInfo :
    
    def __init__(self,name,age,gpa,gender):

        self.name = name
        self.age = age
        self.gpa = gpa
        self.gender = gender
    
    
    def student_info(self):
        print(self.name)
        print(self.age)
        print(self.gpa)
        print(self.gender)
        
```

```python
s1 = StudentInfo("Beshoy",21,3.0,"male")
s1.student_info()
```

```python
class Calc :
    def __init__(self , num1 : int , num2 : int):
        self.num1 = num1
        self.num2 = num2
        
    def sum (self):
        return self.num1+self.num2
        
    def sub (self):
        return self.num1-self.num2
    
c1 = Calc(1,2)
c1.sub()
```

```python
from abc import ABC,abstractmethod

class Animal (ABC) :
    
    """ ABSTRACT -> MUST HAVE IMPLEMENTATION FOR ALL FUNCTIONS """
    
    @abstractmethod
    def move (self):
        pass
    
    
class Bird (Animal) :
    def move(self):
        print("Bird")
        
        
class Cat (Animal) :
    def move(self):
        print("Cat")

    
b = Bird()
b.move()

c = Cat()
c.move()
```
